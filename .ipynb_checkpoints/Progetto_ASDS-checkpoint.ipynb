```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
import iminuit
from iminuit import Minuit
```

```python
file = 'VarAstro_Star_61740_Obs_110560_2025-10-20.txt'
transit_data = np.loadtxt(file, skiprows=14)
#print(transit_data)
#print(np.shape(transit_data))
```

Link stella: https://simbad.cds.unistra.fr/simbad/sim-basic?Ident=HD189733&submit=SIMBAD+search

Link pianeta: https://exoplanet.eu/catalog/hd_189733_ab--307/, https://exoplanetarchive.ipac.caltech.edu/overview/HD%20189733

```python
time = transit_data[:, 0]
mag = transit_data[:, 1]

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(time, mag)
ax.invert_yaxis()

plt.show()
```

```python
df = pd.DataFrame(transit_data)
df.columns=('time', 'mag')
df
```

```python
df['delta'] = df.loc[:, 'time'] - df.at[0, 'time']
df
```

```python
delta = df['delta']

m_media_p = df.loc[0:10, 'mag'].mean() #mag media dei primi 10 pt
m_media_u = df.loc[474:484, 'mag'].mean() #mag media degli ultimi 10 pt

d_media_p = df.loc[0:10, 'delta'].mean() #delta tempi media dei primi 10 pt
d_media_u = df.loc[474:484, 'delta'].mean() #delta tempi media degli ultimi 10 pt

m = (m_media_p-m_media_u)/(d_media_p-d_media_u)
b =(d_media_p*m_media_u - d_media_u*m_media_p)/(d_media_p-d_media_u)
y = m*delta + b
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(delta, mag, c='c')
ax.plot(delta, y, c='y')
ax.invert_yaxis()

plt.show()
```

```python
mag_new = df.loc[:, 'mag'] - m*df.loc[:, 'delta']
plt.plot(df.loc[:, 'delta'], m*df.loc[:, 'delta'])
plt.gca().invert_yaxis()
plt.show()

#questa  la correzione alle y (cio alla magnitudine) che stiamo facendo
```

```python
df['mag_new']= mag_new

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(delta, mag, c='c')
ax.scatter(delta, mag_new, c='b')
ax.plot(delta, y, c='y')
ax.invert_yaxis()

n_m_media_p = df.loc[0:10, 'mag_new'].mean() #rifacciamo le medie per le magnitudini
n_m_media_u = df.loc[474:484, 'mag_new'].mean()

#n_d_media_p = df.loc[0:10, 'delta'].mean()
#n_d_media_u = df.loc[474:484, 'delta'].mean()

n_m = (n_m_media_p-n_m_media_u)/(d_media_p-d_media_u)
n_b =(d_media_p*n_m_media_u - d_media_u*n_m_media_p)/(d_media_p-d_media_u)
n_y = n_m*delta + n_b

ax.plot(delta, n_y, c='r')

plt.show()

print(m, b)
print(n_m, n_b)
```

```python
#METODO ALTERNATIVO
from iminuit.cost import LeastSquares

def line(x, m, b):
    return m*x + b

mag_p10 = df.loc[0:9, 'mag']
mag_u10 = df.loc[475:, 'mag']

delta_p10 = df.loc[0:9, 'delta']
delta_u10 = df.loc[475:, 'delta']

mag10 = np.append(mag_p10, mag_u10)
delta10 = np.append(delta_p10, delta_u10)

sigma_mag10 = np.ones_like(mag10)
leastsquares10 = LeastSquares(delta10, mag10, sigma_mag10, line)

fit10 = Minuit(leastsquares10, m=0.0, b=0.0)
fit10.migrad()
fit10.hesse()

print(fit10.values)
print(fit10.errors)

#farlo con Minuit ha pi senso; quello che noi stiamo facendo  banalmente un fit lineare per raddrizzare i dati
#siccome non abbiamo idea sugli errori, prendiamo come punti solo i primi e gli ultimi 10 punti
#perch prendendo pi punti saranno influenzati dal fatto che noi non abbiamo (ancora) idea di quale sia la forma funzionale
```

```python
mfit10 = fit10.values['m']
bfit10 = fit10.values['b']

mag_fit = df.loc[:, 'mag'] - mfit10*df.loc[:, 'delta']
df['mag_fit']= mag_fit

df
```

```python
fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(12, 4))

ax1.scatter(delta, mag, c='c')
ax1.scatter(delta, mag_fit, c='b')
ax1.plot(delta, y, c='y')
ax1.invert_yaxis()

ax2.scatter(df.index, mag-mag_fit) #l'unica cosa che non mi spiego  il gap

plt.tight_layout()
plt.show()
```

```python
#NOTA: l'unit di misura sulle x  il giorno perch il tempo  espresso in JD (Julian Day)
#Per transformarlo in secondi va moltiplicato per 86400

delta_s = df.loc[:, 'delta']*86400
df['delta_s']= delta_s

df
```

***

$\displaystyle T_\text{dur} = \frac{P}{\pi} \arcsin{\left(\frac{\sqrt{(R_* + R_P)^2 - a^2 \cos{i}^2}}{a}\right)}$

$P$  il periodo, $i$ angolo di inclinazione, $a$  il semiasse maggiore

***

Per $a \gg R_* \gg R_P$, $\displaystyle T_\text{dur} \simeq \frac{P}{\pi} \sqrt{\left(\frac{R_*}{a}\right)^2 -\cos{i}^2}$

Sapendo il tipo di stella, $M_*, R_*$ sono ricavabili dal diagramma HR

$T_\text{dur}, P$ sufficienti a dedurre l'angolo

La transit depth da rapporto tra i raggi: $\frac{\Delta F}{F} = \frac{R_P^2}{R_*^2}$

```python
#Conversione magnitudine-flusso: chiesto dati a Vittorio Tortorella
file_fluxes = 'curva_luce_differenziale.txt'
fluxes = np.loadtxt(file_fluxes, skiprows=1)
print(fluxes)
print(np.shape(fluxes))
```
