```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
file = 'VarAstro_Star_61740_Obs_110560_2025-10-20.txt'
transit_data = np.loadtxt(file, skiprows=14)
#print(transit_data)
#print(np.shape(transit_data))
```

Link stella: https://simbad.cds.unistra.fr/simbad/sim-basic?Ident=HD189733&submit=SIMBAD+search

Link pianeta: https://exoplanet.eu/catalog/hd_189733_ab--307/, https://exoplanetarchive.ipac.caltech.edu/overview/HD%20189733

```python
time = transit_data[:, 0]
mag = transit_data[:, 1]

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(time, mag)
ax.invert_yaxis()

plt.show()
```

```python
df = pd.DataFrame(transit_data)
df.columns=('time', 'mag')
df
```

```python
df['delta'] = df.loc[:, 'time'] - df.at[0, 'time']
df
```

```python
delta = df['delta']

m_media_p = df.loc[0:10, 'mag'].mean() #mag media dei primi 10 pt
m_media_u = df.loc[474:484, 'mag'].mean() #mag media degli ultimi 10 pt

d_media_p = df.loc[0:10, 'delta'].mean() #delta tempi media dei primi 10 pt
d_media_u = df.loc[474:484, 'delta'].mean() #delta tempi media degli ultimi 10 pt

m = (m_media_p-m_media_u)/(d_media_p-d_media_u)
b =(d_media_p*m_media_u - d_media_u*m_media_p)/(d_media_p-d_media_u)
y = m*delta + b
```

```python
fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(delta, mag, c='c')
ax.plot(delta, y, c='y')
ax.invert_yaxis()

plt.show()
```

```python
mag_new = df.loc[:, 'mag'] - m*df.loc[:, 'delta']
plt.plot(df.loc[:, 'delta'], m*df.loc[:, 'delta'])
plt.gca().invert_yaxis()
plt.show()

#questa  la correzione alle y (cio alla magnitudine) che stiamo facendo
```

```python
df['mag_new']= mag_new

fig, ax = plt.subplots(figsize=(12, 8))
ax.scatter(delta, mag, c='c')
ax.scatter(delta, mag_new, c='b')
ax.plot(delta, y, c='y')
ax.invert_yaxis()

n_m_media_p = df.loc[0:10, 'mag_new'].mean() #rifacciamo le medie per le magnitudini
n_m_media_u = df.loc[474:484, 'mag_new'].mean()

#n_d_media_p = df.loc[0:10, 'delta'].mean()
#n_d_media_u = df.loc[474:484, 'delta'].mean()

n_m = (n_m_media_p-n_m_media_u)/(d_media_p-d_media_u)
n_b =(d_media_p*n_m_media_u - d_media_u*n_m_media_p)/(d_media_p-d_media_u)
n_y = n_m*delta + n_b

ax.plot(delta, n_y, c='r')

plt.show()
```

$\displaystyle T_\text{dur} = \frac{P}{\pi} \arcsin{\left(\frac{\sqrt{(R_* + R_P)^2 - a^2 \cos{i}^2}}{a}\right)}$

$P$  il periodo, $i$ angolo di inclinazione, $a$  il semiasse maggiore

***

Per $a \gg R_* \gg R_P$, $\displaystyle T_\text{dur} \simeq \frac{P}{\pi} \sqrt{\left(\frac{R_*}{a}\right)^2 -\cos{i}^2}$

Sapendo il tipo di stella, $M_*, R_*$ sono ricavabili dal diagramma HR

$T_\text{dur}, P$ sufficienti a dedurre l'angolo

La transit depth da rapporto tra i raggi: $\frac{\Delta F}{F} = \frac{R_P^2}{R_*^2}$

Noi abbiamo $N$ punti $(x_i,\ y_i)$; il modello per la dipendenza  del tipo $y = f(x;\ \vec{\theta})$ dove $\vec{\theta} = (\theta_1, \dots, \theta_m)$ sono i nostri parametri

Possiamo assumere che le misure $y_i$ hanno una distribuzione gaussiana attorno a $f(x_i;\ \vec{\theta})$: allora la likelihood $L(\vec{y};\ \vec{\theta})$  il prodotto di $N$ gaussiane e massimizzare quest'ultima corrisponde a minimizzare $-2\log L(\vec{y};\ \vec{\theta})$
